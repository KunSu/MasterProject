```python
pop = pd.read_csv(sample_data_path + 'popular_jobs.csv')
```

```python
pop
```

```python
# revise the name
pop.rename(columns={' JobIds':'JobIds'}, inplace=True)
pop.dropna()
```

```python
# remove the ones that is not a string
pop['rating'] = pop['JobIds'].apply(lambda x: type(x) is str)
pop = pop.loc[pop['rating'], :]
```

```python
pop['JobIds'] = pop['JobIds'].apply(lambda x: x.split())
#pop['JobIds'].apply(lambda x: x.split())
```

```python
def getTOP10(x):
    if len(x)>10:
        x = x[:10]
    return x
pop['JobIds'] = pop['JobIds'].apply(lambda x: getTOP10(x))
#pop['JobIds'].apply(lambda x: getTOP10(x))
pop['rating'] = pop['rating'].replace(to_replace =True, 
                 value =1) 
```

```python
pop['JobIds']
```

```python
lst_col = 'JobIds'

r = pd.DataFrame({
      col:np.repeat(pop[col].values, pop[lst_col].str.len())
      for col in pop.columns.drop(lst_col)}
    ).assign(**{lst_col:np.concatenate(pop[lst_col].values)})[pop.columns]
```

```python
r
```

```python
#Final - get the user_profile_ratings
r.to_csv(sample_data_path + 'pop_rating.csv')
```
